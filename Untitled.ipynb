```python

def SimPearson(i,k):
    moyennei=0
    moyennej=0

    #moyenne des notes de i Ã  j
    x=0
    for j in range(DataFrame.shape[1]):
        if (DataFrame.iloc[i,j]!=-1 and DataFrame.iloc[k,j]!=-1):
            moyennei=moyennei+DataFrame.iloc[i,j]
            x=x+1
    moyennei=moyennei/x

    #moyenne des notes de k Ã  j
    x=0
    for j in range(DataFrame.shape[1]):
        if(DataFrame.iloc[k,j]!=-1 and DataFrame.iloc[i,j]!=-1):
            moyennej=moyennej+DataFrame.iloc[k,j]
            x=x+1
    moyennej=moyennej/x


    Somme =0
    for j in range (DataFrame.shape[1]):
        if(DataFrame.iloc[k,j]!=-1 and DataFrame.iloc[i,j]!=-1):
            Somme=Somme + (DataFrame.iloc[i,j]-moyennei)*(DataFrame.iloc[k,j]-moyennej)


    s1=0
    s2=0
    for j in range (DataFrame.shape[1]):
        if(DataFrame.iloc[k,j]!=-1 and DataFrame.iloc[i,j]!=-1):
            s1=s1+(DataFrame.iloc[i,j]-moyennei)**2

    for j in range (DataFrame.shape[1]):
        if(DataFrame.iloc[k,j]!=-1 and DataFrame.iloc[i,j]!=-1):
            s2=s2+(DataFrame.iloc[k,j]-moyennej)**2

    denominateur= np.sqrt(s1*s2)
    simPearson=Somme/denominateur
    return simPearson
print(SimPearson(0,99))
```

```python
#COSINUS

def SimCos(i, k) :
    Somme = 0
    for j in range(DataFrame.shape[1]) :
            if(DataFrame.iloc[i,j] != -1 and DataFrame.iloc[k,j] != -1) :
                Somme = Somme + DataFrame.iloc[0,j] * DataFrame.iloc[k,j]

    Sommeij = 0
    Sommekj = 0
    for j in range(DataFrame.shape[1]) :
        if(DataFrame.iloc[i,j] != -1 and DataFrame.iloc[k,j] != -1) :
            Sommeij = Sommeij + (DataFrame.iloc[0,j])**2

    for j in range(DataFrame.shape[1]) :
         if(DataFrame.iloc[i,j] != -1 and DataFrame.iloc[k,j] != -1) :
            Sommekj = Sommekj + (DataFrame.iloc[k,j])**2

    denominateur = np.sqrt(Sommeij * Sommekj)
    CosineSim = Somme / denominateur
    return CosineSim
print(SimCos(0,99))
```

```python
#Calcul des similaritÃ©s
SimilaritesPearson= np.zeros(shape=(100,100))
for i in range(100):
    for j in range(100):
        SimilaritesPearson[i][j]=SimPearson(i,j)
print(SimilaritesPearson)
```
