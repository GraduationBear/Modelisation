```python
#CALCUL DES 10 premiÃ¨res similaritÃ©s cosinus en 0 et (1,10)
listeSimCos = []
for i in range(1,10):
    Somme =0
    for j in range (0,DataFrame.shape[1]-1):
        if (DataFrame.iloc[0,j]!=-1 and DataFrame.iloc[i,j]!=-1):
            Somme=Somme + DataFrame.iloc[0,j]*DataFrame.iloc[i,j]

    s1=0
    s2=0
    for j in range (0,DataFrame.shape[1]-1):
        if (DataFrame.iloc[0,j]!=-1 and DataFrame.iloc[i,j]!=-1):
            s1=s1+(DataFrame.iloc[0,j])**2

    for j in range (0,DataFrame.shape[1]-1):
        if(DataFrame.iloc[i,j]!=-1 and DataFrame.iloc[0,j]!=-1):
            s2=s2+(DataFrame.iloc[i,j])**2

    denominateur= np.sqrt(s1*s2)
    simCos=Somme/denominateur
    listeSimCos.append(simCos)
print(listeSimCos)
```

```python
#CALCUL DES 10 premiÃ¨res similaritÃ©s pearson en 0 et (1,10)
listeSimPearson = []
for i in range(1,10):
    moyennei=0
    moyennej=0

    #moyenne des notes de i Ã  j
    x=0
    for j in range(0,DataFrame.shape[1]-1):
        if (DataFrame.iloc[0,j]!=-1 and DataFrame.iloc[i,j]!=-1):
            moyennei=moyennei+DataFrame.iloc[0,j]
            x=x+1
    moyennei=moyennei/x

    #moyenne des notes de k Ã  j
    x=0
    for j in range(0,DataFrame.shape[1]-1):
        if(DataFrame.iloc[i,j]!=-1 and DataFrame.iloc[0,j]!=-1):
            moyennej=moyennej+DataFrame.iloc[i,j]
            x=x+1
    moyennej=moyennej/x


    Somme =0
    for j in range (0,DataFrame.shape[1]-1):
        if(DataFrame.iloc[i,j]!=-1 and DataFrame.iloc[0,j]!=-1):
            Somme=Somme + (DataFrame.iloc[0,j]-moyennei)*(DataFrame.iloc[i,j]-moyennej)


    s1=0
    s2=0
    for j in range (0,DataFrame.shape[1]-1):
        if(DataFrame.iloc[i,j]!=-1 and DataFrame.iloc[0,j]!=-1):
            s1=s1+(DataFrame.iloc[0,j]-moyennei)**2

    for j in range (0,DataFrame.shape[1]-1):
        if(DataFrame.iloc[i,j]!=-1 and DataFrame.iloc[0,j]!=-1):
            s2=s2+(DataFrame.iloc[i,j]-moyennej)**2

    denominateur= np.sqrt(s1*s2)
    simPearson=Somme/denominateur
    listeSimPearson.append(simPearson)
print(listeSimPearson)
```
